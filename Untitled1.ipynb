```python
from sklearn.tree import DecisionTreeRegressor

from sklearn.model_selection import GridSearchCV

from sklearn.datasets import load_diabetes

from sklearn import tree

from sklearn.metrics import mean_squared_error

import matplotlib.pyplot as plt

from sklearn.model_selection import KFold

from sklearn.model_selection import cross_val_score
```

```python
# 加载糖尿病数据集
diabetes = load_diabetes()

X = diabetes.data

y = diabetes.target
```

```python
# 定义决策树回归模型

regressor = DecisionTreeRegressor()
```

```python
# 定义超参数网格
param_grid = {    
    
    'max_depth': [2, 3, 4, 5, 6],   
    
    'min_samples_split': [2, 3, 4, 5, 6]
    
}
```

```python
# 创建交叉验证

kf = KFold (n_splits=10, shuffle=True, random_state=42)
```

```python
# 进行网格搜索

grid_search = GridSearchCV(
                           regressor,
                           
                           param_grid,
                           
                           cv=kf,
                           
                           scoring='neg_mean_squared_error'
                          )

grid_search.fit(X,y)
```

```python
# 获取最优模型

best_regressor = grid_search.best_estimator_
```

```python
# 打印最优超参数

print("最优超参数：", grid_search.best_params_)
```

```python
# 取 10 折交叉验证的 MSE 得分

mse_scores = cross_val_score(
    
    best_regressor, 
    
    X, 
    
    y, 
    
    cv=kf, 
    
    scoring='neg_mean_squared_error')

mse_scores = -mse_scores  # 将负的 MSE 转换为正
```

```python
print ("最优超参数对应的平均 MSE:", mse_scores.mean())
```

```python
# 绘制回归树

plt.figure(figsize=(20, 10))

tree.plot_tree(best_regressor, 
               
               feature_names=diabetes.feature_names, 
               
               filled=True)

plt.show()
```

```python
import shap

explainer = shap.TreeExplainer(best_regressor)

# 计算shap值为numpy.array数组

shap_values_numpy = explainer.shap_values(X)

# 计算shap值为Explanation格式

shap_values_Explanation = explainer(X)
```

```python
shap.plots.beeswarm(shap_values_Explanation, show=False,max_display=13)
```

```python
# 以SHAP的Explanation对象形式输出SHAP值

shap_obj = explainer(X)
```

```python
# 基于模型创建可解释器

explainer = shap.Explainer(best_regressor)

# 基于解释器对特征矩阵X进行解释，算出shap值

shap_values = explainer(X) 
```
