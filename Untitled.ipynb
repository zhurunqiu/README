```python
# 导入sklearn中的datasets模块，用于获取示例数据集
from sklearn import datasets

# 导入numpy库，用于数值计算
import numpy as np

# 导入pandas库，用于数据处理
import pandas as pd

# 导入SHAP库，用于解释模型
import shap

# 导入matplotlib.pyplot模块，用于绘图
import matplotlib.pyplot as plt

# 导入 RandomForestClassifier模块，用于构建随机森林分类器模型
from sklearn.ensemble import RandomForestClassifier

# 导入train_test_split模块，用于划分训练集和测试集
from sklearn.model_selection import train_test_split

# 导入classification_report模块，用于生成分类报告
from sklearn.metrics import classification_report

# 导入seaborn库，用于绘制统计图表
import seaborn as sns

# 导入MinMaxScaler模块，用于数据归一化处理
from sklearn.preprocessing import MinMaxScaler

#导入mean_absolute_error模块，用于计算平均绝对误差
from sklearn.metrics import mean_absolute_error

# 设置字体为新罗马数字
plt.rcParams['font.family'] = 'Times New Roman'

# 处理负号          
plt.rcParams['axes.unicode_minus'] = False 

# 忽略警告
import warnings
          
warnings.filterwarnings("ignore")
```

```python
# 乳腺癌——二分类问题
cancer = datasets.load_breast_cancer() # 从sklearn的datasets模块加载乳腺癌数据集
```

```python
# 提取数据和目标值
data = cancer.data    # 获取数据集中的特征数据
target = cancer.target # 获取数据集中的目标值
```

```python
# 提取特征名称
feature = cancer.feature_names # 获取特征名称
```

```python
# 将数据转换为DataFrame格式
data = pd.DataFrame(data,columns=feature) # 创建一个DataFrame，列名为特征名称
```

```python
# 添加目标列
data['target'] = target # 将目标值作为新列添加到DataFrame中
```

```python
# 将DataFrame保存为CSV文件
data.to_csv('breast_cancer_dataset.csv', index=False) # 将DataFrame保存为CSV文件，不保存行索引
```

```python
import pandas as pd

# 读取CSV文件
df = pd.read_csv('breast_cancer_dataset.csv')

# 获取特征名称

feature = df.columns.tolist()
```

```python
print(feature)
```

```python
# 删除 'target' 列

X = df.drop(['target'], axis=1)

# 获取 'target' 列作为目标变量

y = df['target']
```

```python
# 数据全部为数值型，直接归一化即可（但是由于基分类器都是树模型，归一化也可不做）
# 对特征进行归一化处理  data[feature] = MinMaxScaler().fit_transform(data[feature])
# 将数据集分为训练集和测试集，测试集比例为20%
train_data, test_data, train_label,test_label = train_test_split(X,
                                                                 y,
                                                                 random_state=42,
                                                                 test_size=0.2)

# 初始化随机森林分类器，直接采用默认参数
rf = RandomForestClassifier()

# 训练随机森林模型
rf_model = rf.fit(train_data,train_label)

# 使用训练好的模型进行预测
rf_res = rf_model.predict(test_data)
```

```python
# 将 train_data 转换为 DataFrame
train_data_df = pd.DataFrame(train_data)

# 导出为 CSV 文件
train_data_df.to_csv('train_data.csv', index=False)
```

```python
# 导入SHAP库

import shap

# 使用shap获得一个rf解释器

explainer_rf = shap.TreeExplainer(rf_model)

# 获取训练集各个样本各个特征的SHAP值

shap_values = explainer_rf.shap_values(train_data[feature])

# 获得基线ybase

ybase_rf = explainer_rf.expected_value

# 打印基线值，二分类问题会有两个基线值

print(ybase_rf)
```

```python
print("Shape of data matrix:", data.shape)
print("Shape of shap_values matrix:", shap_values.shape)
```

```python
# 类似于特征重要度，但是衡量指标为shap_value（即x轴），此处仅为负样本的情况

# 使用SHAP库绘制特征重要度的总结图（柱状图类型）

# shap_values[1] 表示第一个类别的SHAP值

shap.summary_plot(shap_values[:,:,1],train_data,plot_type='bar')
```

```python
# 类似于特征重要度，但是衡量指标为shap_value（即x轴），此处为全体样本的情况
# 使用SHAP库绘制特征重要度的总结图
# 这里的shap_values表示所有类别的SHAP值
shap.summary_plot(shap_values[:,:,1],train_data)
```

```python
# 使用SHAP库绘制特定特征的依赖图
# "mean radius" 表示特征名称
# shap_values[0] 表示第一个类别的SHAP值
# train_data 表示训练集数据
# display_features=train_data 用于显示特征名称
shap.dependence_plot("mean radius", shap_values[:,:,0],train_data,display_features=train_data)
```

```python
# 使用SHAP库绘制特定特征的依赖图，不考虑与其他特征的交互作用
# "mean texture" 表示特征名称
# shap_values[1] 表示第一个类别的SHAP值
# train_data 表示训练集数据
# interaction_index=None 表示不考虑与其他特征的交互作用
# show=False 表示不显示图像，只生成图像对象
shap.dependence_plot('mean texture',shap_values[:,:,1],train_data,interaction_index=None,show=False)
```

```python
# 导入 SHAP 库
import shap

# 初始化 SHAP 的 JavaScript 环境

shap.initjs()

# 绘制力导向图，红色表示正向作用，蓝色表示反向作用

# explainer_rf.expected_value[1] 表示第一个类别的基线值

# shap_values[1][3] 表示第一个类别的第四个样本的 SHAP 值

# train_data.iloc[3,:] 表示训练集中的第四个样本

shap.force_plot(explainer_rf.expected_value[1],shap_values[:,:,1][3],train_data.iloc[3,:])
```

```python
# 和上图一样，只不过这是全体都进行了描述。可以选择x轴和y轴的具体定义
# 绘制所有样本的SHAP值力图
# explainer_rf.expected_value[1] 表示第一个类别的基线值
# shap_values[1] 表示第一个类别的所有样本的SHAP值
# train_data[feature] 表示训练集数据中的所有特征
shap.force_plot(explainer_rf.expected_value[1],shap_values[:,:,1],train_data[feature])
```
